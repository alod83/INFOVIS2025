# Radar chart

```python
import matplotlib.pyplot as plt
import numpy as np

def build_radar(human_values, ai_values, title):

    # Define categories of the survey
    categories = [
        "Climate Attitudes",
        "Beliefs & Agency",
        "Social Motivation",
        "Perceived Barriers",
        "Behavioral Intent"
    ]

    # Number of categories
    N = len(categories)


    # Repeat the first value at the end to close the circle
    human_values += human_values[:1]
    ai_values += ai_values[:1]

    # Compute angle for each category
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
    angles += angles[:1]

    # Initialize the radar chart
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

    # Plot human data
    ax.plot(angles, human_values, linewidth=2, linestyle='solid', label="Human", color='orange')
    ax.fill(angles, human_values, alpha=0.25, color='#FFEABF')

    # Plot AI data
    ax.plot(angles, ai_values, linewidth=2, linestyle='dashed', label="AI (ChatGPT-4)", color='red')
    ax.fill(angles, ai_values, alpha=0.25, color='#ff9e81')

    # Set category labels
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=14)

    # Customize radial axis
    ax.set_rlabel_position(30)
    ax.set_yticks([2, 3, 4, 5])
    ax.set_yticklabels(["2", "3", "4", "5"], fontsize=12)
    ax.set_ylim(1, 5)

    # Add title and legend
    plt.title(title, size=20, pad=50)
    plt.legend(loc='lower right', bbox_to_anchor=(1.2, 0.0),fontsize=12)

    # Display the chart
    plt.tight_layout()
    plt.show()
```

```python
# skeptical but attentive
# Simulated average Likert-scale scores for each group
human_values = [3.51, 3.28, 3.25, 3.18, 3.27 ]
ai_values = [2.4, 2.13, 2.4, 2.2, 1.92]
title="Skeptical but Attentive"
build_radar(human_values, ai_values, title)
```

# Heatmap

```python
import seaborn as sns
import pandas as pd


def plot_heatmap(df):
    # Crea le heatmap
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Heatmap per Pearson correlation
    sns.heatmap(df[["Pearson correlation"]], annot=True, cmap="coolwarm", center=0, ax=axes[0], cbar=True,annot_kws={"size": 15})
    axes[0].set_title("Pearson Correlation",fontsize=20)
    axes[0].set_ylabel("")

    # Heatmap per MAE
    sns.heatmap(df[["MAE"]], annot=True, cmap="YlGnBu", ax=axes[1], cbar=True, yticklabels=False, annot_kws={"size": 15})
    axes[1].set_title("Mean Absolute Error (MAE)",fontsize=20)
    axes[1].set_ylabel("")

    axes[0].tick_params(axis='both', labelsize=15)
    axes[1].tick_params(axis='both', labelsize=15)

    # Layout ordinato
    plt.tight_layout()
    plt.show()
```

```python
# skeptical but attentive
data = {
    "Group of questions": [
        "Climate Attitudes and Awareness",
        "Beliefs and Perceived Agency",
        "Social Expectations and Motivation",
        "Perceived Barriers",
        "Behavioral Intent and Engagement"
    ],
    "Pearson correlation": [0.99, -0.99, 0.22, 0.46, 0.34],
    "MAE": [1.12, 0.98, 0.94, 0.50, 0.52]
}

# Creazione DataFrame
df = pd.DataFrame(data)

# Imposta i gruppi come indice
df.set_index("Group of questions", inplace=True)

plot_heatmap(df)
```
