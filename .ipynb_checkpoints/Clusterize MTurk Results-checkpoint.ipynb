```python
# import the required libraries
# load the source/pilot_data.csv file as a pandas dataframe

import pandas as pd

df = pd.read_csv('source/survey_data.csv')

df.head()
```

```python
# drop Q1 Q2 Q3 columns

df.drop(['Q1', 'Q2', 'Q3'], axis=1, inplace=True)
df.dropna(inplace=True)
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=4,init='k-means++',)

model = model.fit(df)
# add the cluster labels to the dataframe
df['cluster'] = model.labels_
df.groupby('cluster').mean().to_csv('output/cluster_means_survey_data_profiles.csv')
```
